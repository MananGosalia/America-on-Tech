```python
# Import the relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preview the dataset
df1 = pd.read_csv('/Users/manan/Downloads/AOT/Module 7/Module7_assignment_dataset.csv')
df1.head()
```

```python
# Calculate mean, median, and mode for BMI
bmi_mean = df1['BMI'].mean()
bmi_median = df1['BMI'].median()
bmi_mode = df1['BMI'].mode()[0] 

print("BMI - Mean: " + str(bmi_mean) + ", Median: " + str(bmi_median) + ", Mode: " + str(bmi_mode))

# Calculate mean, median, and mode for PhysicalHealth 
physical_mean = df1['PhysicalHealth'].mean()
physical_median = df1['PhysicalHealth'].median()
physical_mode = df1['PhysicalHealth'].mode()[0]

print("Physical Health - Mean: " + str(physical_mean) + ", Median: " + str(physical_median) + ", Mode: " + str(physical_mode))

# Calculate mean, median, and mode for MentalHealth
mental_mean = df1['MentalHealth'].mean()
mental_median = df1['MentalHealth'].median()
mental_mode = df1['MentalHealth'].mode()[0]

print("Mental Health - Mean: " + str(mental_mean) + ", Median: " + str(mental_median) + ", Mode: " + str(mental_mode))

# Calculate mean, median, and mode for SleepTime
sleep_mean = df1['SleepTime'].mean()
sleep_median = df1['SleepTime'].median()
sleep_mode = df1['SleepTime'].mode()[0]

print("Sleep Time - Mean: " + str(sleep_mean) + ", Median: " + str(sleep_median) + ", Mode: " + str(sleep_mode))
```

```python
# Calculate standard deviation and variance for SleepTime
sleep_sd = df1['SleepTime'].std()
sleep_variance = df1['SleepTime'].var()

print("Sleep Time - Standard Deviation: " + str(sleep_sd) + ", Variance: " + str(sleep_variance))

# Calculate standard deviation and variance for BMI
bmi_sd = df1['BMI'].std()
bmi_variance = df1['BMI'].var()

print("BMI - Standard Deviation: " + str(bmi_sd) + ", Variance: " + str(bmi_variance))

# The BMI data is positively skewed, because the right side of the distribution is longer than the left. 
```

```python
heart_disease_counts = df1['HeartDisease'].value_counts()

# Create a pie chart
plt.figure(figsize=(4, 4))
plt.pie(heart_disease_counts, labels=heart_disease_counts.index, autopct='%1.1f%%', colors=['#FFCCCB', '#BFEDD4'])
plt.title("Proportion of Individuals with Heart Disease")
plt.show()
```

```python
# Histogram for BMI
plt.figure(figsize=(4, 4))
sns.histplot(df1['BMI'], bins=10, color='blue')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```
