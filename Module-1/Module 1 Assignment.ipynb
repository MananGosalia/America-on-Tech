```python
# Import the pandasql library
from pandasql import sqldf
```

```python
# Run SQL query to select name, email, and loyalty_status from the appended files
sqldf(
"""
        SELECT first_name, last_name, email, loyalty_status 
        FROM df_append;
"""
     )
```

```python
# Run SQL query to select customers that have loyalty_status of 'Gold'
sqldf(
"""
        SELECT first_name, last_name, email, loyalty_status 
        FROM df_append 
        WHERE loyalty_status = 'Gold';
"""
     )
```

```python
# Run SQL query to select first name, email, total spend, and most traveled state
sqldf(
"""
        SELECT first_name, email, total_spend_2021, most_traveled_state 
        FROM df_append;
"""
     )
```

```python
# Run SQL query to select loyalty status, education, and total spend for customers from NY
sqldf(
"""
        SELECT loyalty_status, education, total_spend_2021
        FROM df_append
        WHERE most_traveled_state = 'NY';
"""
     )
```
