```````````````````````````python
# Create correlation heatmap for numerical features
plt.figure(figsize=(12, 8))
numerical_cols = ['age', 'gpa', 'Perseverance', 'DesireToTakeInitiative',
                 'Competitiveness', 'SelfReliance', 'StrongNeedToAchieve',
                 'SelfConfidence', 'GoodPhysicalHealth']
correlation_matrix = df[numerical_cols].corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create heatmap
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm',
            center=0, fmt='.2f', square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()

# Create pairplot for key features colored by StartedBusiness
key_features = ['age', 'gpa', 'Perseverance', 'SelfConfidence', 'StartedBusiness']
sns.pairplot(df[key_features], hue='StartedBusiness', diag_kind='kde')
plt.suptitle('Pairwise Relationships between Key Features', y=1.02)
plt.show()

# Create violin plots to compare distributions between entrepreneurs and non-entrepreneurs
plt.figure(figsize=(15, 6))
personality_traits = ['Perseverance', 'DesireToTakeInitiative', 'Competitiveness',
                     'SelfReliance', 'StrongNeedToAchieve', 'SelfConfidence']

# Reshape data for violin plots
melted_df = pd.melt(df, id_vars=['StartedBusiness'], value_vars=personality_traits,
                    var_name='Trait', value_name='Score')

# Create violin plot
sns.violinplot(data=melted_df, x='Trait', y='Score', hue='StartedBusiness')
plt.xticks(rotation=45)
plt.title('Distribution of Personality Traits by Entrepreneurship Status')
plt.tight_layout()
plt.show()

# Create boxplots for GPA by major and entrepreneurship status
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='major', y='gpa', hue='StartedBusiness')
plt.xticks(rotation=45)
plt.title('GPA Distribution by Major and Entrepreneurship Status')
plt.tight_layout()
plt.show()

# Create a custom function for calculating percentages
def calculate_business_percentage(data, column):
    return (data[data['StartedBusiness'] == 1][column].value_counts() /
            data[column].value_counts() * 100).sort_values(ascending=False)

# Calculate and plot business startup rates by major
business_rates_major = calculate_business_percentage(df, 'major')
plt.figure(figsize=(10, 6))
business_rates_major.plot(kind='bar')
plt.title('Percentage of Students Who Started Business by Major')
plt.xlabel('Major')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create stacked bar chart for mentor influence
plt.figure(figsize=(8, 6))
mentor_business = pd.crosstab(df['HasMentor'], df['StartedBusiness'], normalize='index') * 100
mentor_business.plot(kind='bar', stacked=True)
plt.title('Impact of Mentorship on Business Creation')
plt.xlabel('Has Mentor')
plt.ylabel('Percentage')
plt.legend(title='Started Business', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```````````````````````````

```````````````````````````python
#We will use IQR method to find outliers in the dataset

import pandas as pd
import numpy as np


# Step 1: Calculate IQR for each column
IQR = df_numeric_without_age_id.quantile(0.75) - df_numeric_without_age_id.quantile(0.25)
Q1 = df_numeric_without_age_id.quantile(0.25)
Q3 = df_numeric_without_age_id.quantile(0.75)
# Step 2: Determine lower and upper bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Step 3: Identify outliers
outliers = [(df_numeric_without_age_id < lower_bound.T).any(axis=1) | (df_numeric_without_age_id > upper_bound.T).any(axis=1)]

df_numeric_with_no_outliers = df_numeric_without_age_id[~((df_numeric_without_age_id < lower_bound) | (df_numeric_without_age_id > upper_bound))]
df_numeric_with_no_outliers.dropna()
```````````````````````````
